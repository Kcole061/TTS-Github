### Exercise 1. 

These data consist of Human Development Index and Corruption Perception Index scores for several countries.

- A. Create a scatter plot with CPI on the x axis and HDI on the y axis.
- B. Color the points blue.
- C. Map the color of the the points to Region.
- D. Make the points bigger by setting size to 2
- E. Map the size of the points to HDI.Rank

````r
# Create a scatter plot with CPI on the x axis and HDI on the y axis.
s1 <- ggplot(eData, aes (x=CPI, y=HDI)) 
s1 + geom_point()
````

````r
#Color the points blue
s1 <- ggplot(eData, aes (x=CPI, y=HDI))
s1+geom_point(color="blue")
````

````r
#Map the color of the the points to Region
s1 <- ggplot(eData, aes (x=CPI, y=HDI))
s1+geom_point(aes(color = Region))
````

````r
# Make the points bigger by setting size to 2
s1 <- ggplot(eData, aes (x=CPI, y=HDI))
s1+geom_point(aes(color = Region),size =2)
````

````r
# Map the size of the points to HDI.Rank
s1 <- ggplot(eData, aes (x=CPI, y=HDI))
s1+geom_point(aes(color = Region),size= HDI.Rank)
````
