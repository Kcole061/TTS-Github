### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd

csd = pd.read_csv('C:/Users/kahli/Desktop/TTS Files/TTS.Student-main/TTS.Student-main/datascience/python/11_mod_matplotlib/data/company_sales_data.csv')
csd.head(10)
````

````python
plt.plot(csd['month_number'], csd['total_profit'])
plt.title("Total Profit of All Months")
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
````

### Exercise 2. 

Get Total Units Sold of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Units Sold Number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
plt.plot(csd['month_number'], csd['total_profit'], linestyle = 'dashed', color ='red', marker='o', lw = 3)
plt.title("Total Profit of All Months")
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')

plt.legend("Total Proftis Over Time")
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
plt.plot(csd['month_number'], csd['facecream'], linestyle = ':', color ='black', lw = 3, marker='o', label = 'Facecream')
plt.plot(csd['month_number'], csd['facewash'], linestyle = '-', color ='red', lw = 3, marker='D', label = 'Facewash')
plt.plot(csd['month_number'], csd['toothpaste'], linestyle = ':', color ='c', lw = 3, marker='H', label = 'toothpaste')
plt.plot(csd['month_number'], csd['bathingsoap'], linestyle = '-.', color ='blue', lw = 3, marker='+', label = 'bathingsoap')
plt.plot(csd['month_number'], csd['shampoo'], linestyle = ':', color ='m', lw = 3, marker='x', label = 'shampoo')
plt.plot(csd['month_number'], csd['moisturizer'], linestyle = ':', color ='g', lw = 3, marker='v', label = 'moisturizer')
plt.xlabel('Month Number')
plt.ylabel('Units Sold Per Month')
plt.legend(loc = 'upper left')
plt.title('Units Sold Over Time')
plt.show()
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
plt.scatter(csd['month_number'], csd['toothpaste'])
plt.xlabel('Month Number')
plt.ylabel('Tooth Paste Sold Units Number')
plt.grid(color='k', linewidth= 1, linestyle="--")
plt.title("Tooth Paste ")
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
plt.bar(csd['month_number']-0.25, csd['facecream'], label = 'Facecream',  align='edge', width= 0.5)
plt.bar(csd['month_number']+0.25, csd['facewash'], label = 'Facewash',  align='edge', width= 0.5)
plt.legend(loc = 'lower right')
plt.xlabel('Month Number')
plt.ylabel('Units Sold Per Month')
plt.title('Facecream and Facewash Units Sold')
````
