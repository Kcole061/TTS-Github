### Exercise 3
From the data, remove rows containing NA’s. Fit the linear model with the formulas from exercise 2.

````r
# exercise 3
cleanHeights <- na.omit(heights)
````

````r
im2 <- lm(height ~ weight, data = cleanHeights)
coef(im2)
````

````r
grid <- cleanHeights %>% 
  data_grid(weight) 
grid
````

````r
grid <- grid %>% add_predictions(im2)
grid
````

````r
ggplot(cleanHeights, aes(weight)) + geom_point(aes(y = height)) +
  geom_line(aes(y = pred), data = grid, color = "red", size = 1)
````

### Exercise 4
For each fit, calculate ```RMSE```.

````r
# exercise 4
sqrt(mean((cleanHeights$weight - grid$pred )^2))
````

### Exercise 5
For each model, add residuals to the data and plot their distribution. (Hint: ```use lift_dl()```.)

````r
# exercise 5
im3 <- cleanHeights %>% add_residuals(im2)
im3
````

````r
ggplot(im3, aes(weight, resid)) + geom_ref_line(h = 0) + geom_point()
````
